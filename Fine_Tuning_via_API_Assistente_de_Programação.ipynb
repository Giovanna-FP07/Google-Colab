```python
from google.colab import userdata
import google.generativeai as genai

CHAVE_API_GOOGLE = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=CHAVE_API_GOOGLE)
MODELO_ESCOLHIDO = "gemini-2.0-flash"
```

```python
def assistente_programacao_gemini(pergunta):

  # Assistente de estudo especializado em programação usando o Google Gemini

  # Configuração do modelo
  llm = genai.GenerativeModel(
      model_name=MODELO_ESCOLHIDO,
      system_instruction="""

      Você é um tutor especializado em programação para estudantes universitários.

      Diretrizes específicas:
      - Explique conceitos de programação de forma clara e passo a passo
      - Forneça exemplos de código em Python sempre que relevante
      - Use analogias do mundo real para explicar conceitos abstratos
      - Corrija erros conceituais de forma educativa e encorajadora
      - Mantenha um tom acessível, mas técnico
      - Enfatize boas práticas e padrões de código limpo

      """
  )

  # Configuração de geração - FT
  configuracao_ft = {
      "temperature": 0.3,
      "top_p": 0.8,
      "top_k": 30,
      "stop_sequences":["\n\n---", "Fim da explicação"],
      "max_output_tokens": 1200
  }

  try:
    # Geração da resposta
    response = llm.generate_content(
      contents=pergunta,
      generation_config=configuracao_ft
    )
    return response.text

  except Exception as e:
    print(f"Erro ao gerar resposta: {e}")
```

```python
# Exemplo de uso
pergunta = "Explique o conceito de recursão em programação com exemplos práticos"
resposta = assistente_programacao_gemini(pergunta)
print(f"Resposta do Gemini:\n {resposta}")
```

```python
# Exemplo de uso
pergunta = "Explique o conceito de pilhas em programação com exemplos práticos"
resposta = assistente_programacao_gemini(pergunta)
print(f"Resposta do Gemini:\n {resposta}")
```
