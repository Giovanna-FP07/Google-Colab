```python
import os
from google.colab import userdata
from openai import OpenAI
```

```python
# Configura a chave da API da OpenAI
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY

# Criação do cliente IA e seleção do modelo
llm = OpenAI()
MODELO_ESCOLHIDO = "gpt-4o-mini"
```

```python
# Lista de categorias possíveis
lista_categorias_possiveis = [
    "Eletrônicos Verdes",
    "Moda Sustentável",
    "Produtos de Limpeza Ecológicos",
    "Alimentos Orgânicos",
    "Higiene Sustentável"
]
```

```python
# Prompt de sistema
prompt_sistema = f"""
Você é um categorizador de produtos.
Você deve assumir as categorias presentes na lista abaixo.

# Lista de Categorias Válidas
{lista_categorias_possiveis}

# Formato da Saída
Produto: Nome do Produto
Categoria: Apresente a Categoria do Produto

# Exemplo de Saída
Produto: Escova elétrica com recarga solar
Categoria: Eletrônicos Verdes
"""
```

```python
# Obtém o prompt do usuário via input
pergunta = input("\n Digite o nome do produto para categorizar: ")

# Envia para a API da OpenAI e obtém o resultado
try:
  response = llm.chat.completions.create(
      model = MODELO_ESCOLHIDO, # ou "gpt-5" se tiver acesso
      messages = [
          {"role": "system", "content": prompt_sistema},
          {"role": "user", "content": pergunta}
      ],
      #temperature = 0.0, # para respostas mais determinísticas
      #max_tokens = 100
  )

  resposta_texto = response.choices[0].message.content.strip()
  print(f"\n✓ Resposta gerada:\n{resposta_texto}")

except Exception as e:
  print(f"\n✗ Erro ao chamar a API: {e}")
```
