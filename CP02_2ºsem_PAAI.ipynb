```python
url = '/content/Guia digital IA - Itaú.pdf'
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(url)
pages = []

for page in loader.lazy_load():
  pages.append(page)
```

```python
print(f'Metadados: {pages[0].metadata}\n')
```

```python
print(f'Resposta: {pages[0].page_content}')
```

```python
print(f'Quantidade de páginas: {len(pages)}')
```

```python
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embed_model = HuggingFaceEmbeddings(model_name="mixedbread-ai/mxbai-embed-large-v1")
```

```python
vector_store = InMemoryVectorStore.from_documents(pages, embed_model)
```

```python
docs = vector_store.similarity_search("Definições de inteligência artificial segundo a OCDE, AI Act e PL 2.338/2023", k=2)
```

```python
for doc in docs:
  print(f'Page {doc.metadata["page"]}: {doc.page_content[:300]} \n')
```

```python
retriever = vector_store.as_retriever()
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
template = """
  You're helpful assistent that only gives answers bases on the given context.
  If the answer is not in the context, say "I don't know".

  Context: {context}
  Question: {question}
  Answer:

"""
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
response = chain.invoke("Quais são as definições de inteligência artificial segundo a OCDE, o AI Act e o PL 2.338/2023?")
print(f'Response: {response}')
```
